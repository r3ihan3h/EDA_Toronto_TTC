<h4 style=' font-family:monospace'><center>How many null do we have in our coulmns?</center></h4>

```python
df.isnull().sum()
```

<h4 style=' font-family:monospace'><center>The Direction Column has many Null values, it is not possible to fill this column and also with respect to the route column we do not need the Direction. But these kind of decisions in most cases depend on our data and what do we need to extract from our data. For now, we just drop this column.</center></h4>

```python
df =df.drop(['Direction'], axis = 1)
```

<h4 style=' font-family:monospace'><center>Next, we have the "Route" column with Null values:</center></h4>


```python
df[(df['Route'].isnull())].head(10)
```

<h4 style=' font-family:monospace'><center>As we can see, most of the columns with Null Routes have Min Delay as 0 also. We do not need 0 minute delay bus records as we are exploring Delay of the buses. So we will remove all the 0 Min Delay rows.</center></h4>


```python
df = df[df['Min Delay'] != 0]
```

<h4 style=' font-family:monospace'><center>The number of Route null values turned to be 26</center></h4>

```python
df.isnull().sum()
```

<h4 style=' font-family:monospace'><center>It Is a small number so we will drop thoes rows also:</center></h4>

```python
df = df[df['Route'].notnull()]
```

```python
df.sample(10)
```

<h4 style=' font-family:monospace'><center>Better to convert Date column to a propper date format</center></h4>

```python
df['Date'] =df['Date'].apply(pd.to_datetime)
```

<h1 style='background:#C4F1E8; border:2; border:8px double yellow;font-size:200%; font-family:Chaparral Pro Light;font-family:Chaparral Pro Light;font-weight: bold;'><center>VISUALIZATION</center></h1>

```python
import cufflinks as cf
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)  #  link pandas to plotly and add the iplot method
```

<h4 style=' font-family:monospace'><center>First we want to know what is the most important reason for Bus's delay in this city? 
So we need to group all the Incidents together and then calculate the sum of the delay:</center></h4>


```python
Incident = df.groupby(df['Incident'])['Min Delay'].sum()/60 # To Hour for making more sence
df_Inc = pd.DataFrame({'labels': Incident.index,
                   'values': Incident.values
                  })
df_Inc.iplot(kind='bar',x = 'labels', y = 'values', title="Delay's Incident", colors = 'palegreen',  #I love this color <3
            xTitle = 'Incident', yTitle = 'Delay in Hours') 
```

<h4 style=' font-family:monospace'><center>The Diversion has the most delay time followed by Operation incidents.
Diversion is a special route arranged for traffic to follow when the normal route cannot be used which dedicated more that *** 2500 *** Hour** only in the first 6 month of the 2022. Can you imagine?
We have Mechanical reason for the delays in the third level wich caused more than 1500 hours of delys.</center></h4>



<h4 style=' font-family:monospace'><center>Now let's see which Routes have the most delay time?
for this purpose we have to group by Routes and sum the delays again.</center></h4>




```python
Route = df.groupby(df['Route'])['Min Delay'].sum().reset_index()
Route_20 = Route.sort_values(by = ['Min Delay'], ascending = False).head(10)
```

<h4 style=' font-family:monospace'><center>These are the top 10 Routes with respect to the delay time. Let's see what is the name of these routes and make them more understandable. I found the name of them in this link : https://www.ttc.ca/routes-and-schedules#/listroutes/bus</center></h4>



```python
Route_20['Route'] = Route_20['Route'] .replace({ '52':'Lawrence West', '36':'Finch West', '32':'Eglinton West', '97':'Yong',
                                      '29':'Dufferin', '96':'Wilson', '54':'Lawrence East', '47':'Lansdowne',
                                      '63':'Ossington', '14':'Glencairn'
                                     })
```

```python
Route_20.iplot(kind='bar',x = 'Route', y = 'Min Delay', title="Top 10 Routes which had the most delay", 
xTitle = 'Route Name', yTitle = 'Delay in Minutes',colors = 'turquoise') #Another Lovely Color
```

<h4 style=' font-family:monospace'><center>We also can investigate about the causes for delay in Lawrence route:</center></h4>

```python
df_lawrence = df.loc[df['Route'] == '54']
df_lawrence_delay = df.groupby(['Incident'])['Min Delay'].sum().reset_index().sort_values(by = ['Min Delay'], ascending = False)
df_lawrence_delay
```

<h4 style=' font-family:monospace'><center>Seems this Route has many problems not one or two. But it probably has too many reasons which is not going to be explore in this Kernel.</center></h4>


<h4 style=' font-family:monospace'><center>Now, lets See what is happening during the week?</center></h4>

```python
Day = df.groupby(df['Day'])['Min Delay'].sum()/60 # To Hour for making more sence
df_Day = pd.DataFrame({'labels': Day.index,
                   'values': Day.values
                  })
df_Day.iplot(kind='pie',labels = 'labels', values = 'values', title="Delay based on day of the week"
             ,pull=[0.02, 0.02,0.02, 0.02,0.02, 0.02,0.02],hole = 0.3)            
```

<h4 style=' font-family:monospace'><center>Seems pretty equal :)</center></h4>


<h4 style=' font-family:monospace'><center>Also we can investigate each weekday based on incidents and the duration of the delay which that incident caused</center></h4>



```python
Monday = df.loc[df['Day']== 'Monday']
Monday_grp = Monday.groupby(['Incident'])['Min Delay'].sum().reset_index()

Saturday = df.loc[df['Day']== 'Saturday']
Saturday_grp = Saturday.groupby(['Incident'])['Min Delay'].sum().reset_index()

Sunday = df.loc[df['Day']== 'Sunday']
Sunday_grp = Sunday.groupby(['Incident'])['Min Delay'].sum().reset_index()

Thursday = df.loc[df['Day']== 'Thursday']
Thursday_grp = Thursday.groupby(['Incident'])['Min Delay'].sum().reset_index()

Tuesday = df.loc[df['Day']== 'Tuesday']
Tuesday_grp = Tuesday.groupby(['Incident'])['Min Delay'].sum().reset_index()

Friday = df.loc[df['Day']== 'Friday']
Friday_grp = Friday.groupby(['Incident'])['Min Delay'].sum().reset_index()

Wednesday = df.loc[df['Day']== 'Wednesday']
Wednesday_grp = Wednesday.groupby(['Incident'])['Min Delay'].sum().reset_index()
```

```python
from plotly.subplots import make_subplots #for making mutiple polts 
import plotly.graph_objects as go

fig = make_subplots(rows=4, cols=2, shared_yaxes=True, vertical_spacing= 0.2,subplot_titles=['Monday','Tuesday',
                                                                                             'Wednesday','Thursday',
                                                                                             'Friday','Saturday',
                                                                                             'Sunday'])

fig.add_trace(go.Bar(x = Monday_grp['Incident'], y = Monday_grp['Min Delay']),
              row=1, col=1)
fig.add_trace(go.Bar(x = Tuesday_grp['Incident'], y = Tuesday_grp['Min Delay']),
              row=1, col=2)
fig.add_trace(go.Bar(x = Wednesday_grp['Incident'], y = Wednesday_grp['Min Delay']),
              row=2, col=1)
fig.add_trace(go.Bar(x = Thursday_grp['Incident'], y = Thursday_grp['Min Delay']),
              row=2, col=2)
fig.add_trace(go.Bar(x = Friday_grp['Incident'], y = Friday_grp['Min Delay']),
              row=3, col=1)
fig.add_trace(go.Bar(x =Saturday_grp['Incident'], y =Saturday_grp['Min Delay']),
              row=3, col=2)
fig.add_trace(go.Bar(x = Sunday_grp['Incident'], y = Sunday_grp['Min Delay']),
              row=4, col=1)


fig.update_layout(height=1500, width=1000,
                  title_text="Week days and their delay time ")
fig.show()
```

<h4 style=' font-family:monospace'><center>Now let's explore the months and see which month had the most delay duration </center></h4>



```python
df_month = df.groupby(df['Date'].dt.month)['Min Delay'].sum()/60  #grouping the delay by month
df_month = df_month.reset_index()
df_month.iplot(kind = 'bar', x= 'Date', y = 'Min Delay', title = 'The total hours of delay in each month', xTitle = 'The Number of the month',
                                    yTitle = 'Hours of delay', color = 'limegreen')
```

<h4 style=' font-family:monospace'><center>What happened in Jan 2022? </center></h4>



```python
df_Jan = df.loc[df['Date'].dt.month == 1]
fig = px.scatter(df_Jan, x="Date", y="Incident", size ="Min Delay")
fig.show()
```
